```python
#19235

score = 0

greenboard = [[0]*4 for _ in range(6)]
blueboard = [[0]*4 for _ in range(6)]
    
def dropblock(ny, board):
    nx = -1
    while 1:
        nx += 1
        if nx == 6:
            nx -= 1
            break
        elif board[nx][ny]:
            nx -= 1
            break
    return nx


def dropblock2(ny, nx, board):
    while 1:
        nx += 1
        if nx == 6:
            nx -= 1
            break
        elif board[nx][ny]:
            nx -= 1
            break
    return nx


def down(board):
    visit = {}
    for h in range(4, -1, -1):
        for w in range(4):
            if visit.get((h, w)) is None and board[h][w]:
                if w < 3 and board[h][w] == board[h][w+1]:
                    min1 = dropblock2(w,h, board)
                    min2 = dropblock2(w+1,h, board)
                    min3 = min(min1,min2)
                    if min3 != h:
                        board[min3][w] = board[min3][w+1] = board[h][w]
                        board[h][w] = board[h][w+1] = 0
                    visit[(h,w)] = visit[(h,w+1)] = 1
                else:
                    visit[(h,w)] = 1
                    idx = dropblock2(w, h, board)
                    if idx != h:
                        board[idx][w] = board[h][w]
                        board[h][w] = 0
                
def scorecheck(board):
    global score
    flag = 0
    for row in range(5, 1, -1):
        cnt = 0
        for w in range(4):
            if board[row][w]:
                cnt += 1
        if cnt == 4:
            score += 1
            for w in range(4):
                board[row][w] = 0
            flag = 1
    if flag:
        down(board)
        scorecheck(board)
        
def cleanboard(board):
    cnt = 0
    if sum(board[0]) > 0:
        cnt += 1
    if sum(board[1]) > 0:
        cnt += 1
    while cnt > 0:
        del board[-1]
        board.insert(0,[0]*4)
        cnt -= 1

n = int(input())

for time in range(1, n+1):
    t, x, y = list(map(int,input().split()))
    if t == 1:
        ng = dropblock(y, greenboard)
        greenboard[ng][y] = time
        nb = dropblock(x, blueboard)
        blueboard[nb][x] = time
    elif t== 2:
        ng1 = dropblock(y, greenboard)
        ng2 = dropblock(y+1, greenboard)
        ng = min(ng1, ng2)
        greenboard[ng][y] = time
        greenboard[ng][y+1] = time
        nb = dropblock(x, blueboard)
        blueboard[nb][x] = time
        nb = dropblock(x, blueboard)
        blueboard[nb][x] = time
    else:
        nb1 = dropblock(x, blueboard)
        nb2 = dropblock(x+1, blueboard)
        nb = min(nb1, nb2)
        blueboard[nb][x] = time
        blueboard[nb][x+1] = time
        ng = dropblock(y, greenboard)
        greenboard[ng][y] = time
        ng = dropblock(y, greenboard)
        greenboard[ng][y] = time
        
    scorecheck(greenboard)
    scorecheck(blueboard)
    cleanboard(greenboard)
    cleanboard(blueboard)
    
b = 0
g = 0
for i in range(6):
    for j in range(4):
        if greenboard[i][j]:
            g += 1
        if blueboard[i][j]:
            b += 1
print(score)
print(b+ g)
```

```python
# #19238
import sys
sys.stdin = open("sample_6.txt",'r')

import copy

global N, M, F
global MATRIX
global TAXI_X, TAXI_Y

  
    
def get_shortest_path(src, dst):
    queue = []
    visited = copy.deepcopy(MATRIX)
    src = [src[0], src[1], 0]
    queue.append(src)
    visited[src[0]][src[1]] = 1

    while len(queue) > 0:
        tmp_src = queue.pop(0)
        
        if (tmp_src[0], tmp_src[1]) == dst:
            return tmp_src[2]
        
        if (tmp_src[0]+1 < N) and (visited[tmp_src[0]+1][tmp_src[1]] != 1):
            queue.append([tmp_src[0]+1, tmp_src[1],tmp_src[2] + 1])
            visited[tmp_src[0]+1][tmp_src[1]] = 1
            
        if (tmp_src[0]-1 >= 0) and (visited[tmp_src[0]-1][tmp_src[1]] != 1):
            queue.append([tmp_src[0]-1, tmp_src[1],tmp_src[2] + 1])
            visited[tmp_src[0]-1][tmp_src[1]] = 1
            
        if (tmp_src[1]+1 < N) and (visited[tmp_src[0]][tmp_src[1]+1] != 1):
            queue.append([tmp_src[0], tmp_src[1]+1,tmp_src[2] + 1])
            visited[tmp_src[0]][tmp_src[1]+1] = 1
            
        if (tmp_src[1]-1 >= 0) and (visited[tmp_src[0]][tmp_src[1]-1] != 1):
            queue.append([tmp_src[0], tmp_src[1]-1,tmp_src[2] + 1])
            visited[tmp_src[0]][tmp_src[1]-1] = 1
    return -1

N, M, F = list(map(int,sys.stdin.readline().split()))
# N, M, F = list(map(int,input().split()))
MATRIX = [list(map(int,sys.stdin.readline().split())) for _ in range(N)]
# MATRIX = [list(map(int,input().split())) for _ in range(N)]
TAXI_X, TAXI_Y= list(map(int,sys.stdin.readline().split()))
# TAXI_X, TAXI_Y = list(map(int,input().split()))
TAXI_X -= 1
TAXI_Y -= 1
passengers = []
for _ in range(M):
    passenger = list(map(int,sys.stdin.readline().split()))
#     passenger = list(map(int,input().split()))
    passengers.append([passenger[0]-1, passenger[1]-1, passenger[2]-1, passenger[3]-1])

done = False
while len(passengers) > 0:
    shortest_dist = float('inf')
    shortest_p = None
    for idx, passenger in enumerate(passengers):
        if (TAXI_X, TAXI_Y) == (passenger[0], passenger[1]):
            shortest_dist = 0
            shortest_p = passenger
            break
        dist = get_shortest_path((TAXI_X, TAXI_Y), (passenger[0], passenger[1]))
        if dist == -1:
            passengers = []
            break
        elif dist < shortest_dist:
            shortest_dist = dist
            shortest_p = passenger
        elif dist == shortest_dist:
            if (passenger[0] <= shortest_p[0]) and (passenger[1] < shortest_p[1]):
                shortest_p = passenger
#     print(F)
#     print(shortest_dist)
#     print(shortest_p)
#     print()
# #     break
    if F >= shortest_dist:
        F -= shortest_dist
        TAXI_X, TAXI_Y = shortest_p[0], shortest_p[1]
        drive_dist = get_shortest_path((TAXI_X, TAXI_Y), (shortest_p[2], shortest_p[3]))
        if drive_dist == -1:
            done = False
            break
        elif F >= drive_dist:
            F += get_shortest_path((TAXI_X, TAXI_Y), (shortest_p[2], shortest_p[3]))
            TAXI_X, TAXI_Y = shortest_p[2], shortest_p[3]
            passengers.remove(shortest_p)
            if len(passengers) == 0:
                done = True
                break
        else:
            done = False
            break

    else:
        done = False
        break
    
    
    
if not done:
    print(-1)
else:
    print(F)
```
