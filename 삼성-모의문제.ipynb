```python
#5658

def get_numbers(string, n):
    return [string[:n], string[n:2*n], string[2*n:3*n], string[3*n:4*n]]

def hex2num(hex_num):
    return int(hex_num, 16)

num_case = int(input())


for case in range(num_case):
    answer = f'#{case+1} '
    N, K = list(map(int,input().split()))
    numbers = input()
    
    num_set = list()
    rotate_num = int(N/4)
    num_set.extend(get_numbers(numbers, rotate_num))
    for i in range(rotate_num):
        numbers = (numbers[-1] + numbers)[:-1]
        num_set.extend(get_numbers(numbers, rotate_num))
    num_set = list(set(num_set))
    num_list = [hex2num(x) for x in num_set]
    num_list.sort(reverse=True)
    print(answer + str(num_list[K-1]))
    
```

```python
# 5656
import copy


global init_sum
global N, W, H

def remove_blocks(x,y, matrix):
    value = matrix[x][y]
    matrix[x][y] = 0
    if value != 1:
        for i in range(1, value):
            if x+i < H:
                remove_blocks(x+i, y, matrix)
            if x-i >= 0:
                remove_blocks(x-i, y, matrix)
            if y+i < W:
                remove_blocks(x, y+i, matrix)
            if y-i >= 0:
                remove_blocks(x, y-i, matrix)

def update_matrix(matrix):# 벽돌을 내려서 빈자리 채움
    for i in range(W):
        tmp = []
        for j in range(H):
            if matrix[j][i] != 0:
                tmp.append(matrix[j][i])
        start = H - len(tmp)
        for k in range(H):
            if k < start:
                matrix[k][i] = 0
            else:
                matrix[k][i] = tmp[k-start]

def get_sum(matrix):
    _sum = 0
    for row in matrix:
        _sum += sum([int(x>0) for x in row])
    return _sum

def get_first_loc(y, matrix):
    x = -1
    for i in range(H):
        if matrix[i][y] != 0:
            x = i
            break
    return x, y

        
def complete_search(n, matrix):
    if n == 0:
        return get_sum(matrix)
    else:
        _min = init_sum
        for j in range(W):
            tmp_matrix = copy.deepcopy(matrix)
            x, y = get_first_loc(j, tmp_matrix)
            if x == -1:
                continue
            remove_blocks(x, y, tmp_matrix)
            update_matrix(tmp_matrix)
            if get_sum(tmp_matrix) == 0:
                _min = 0
                break
            tmp_value = complete_search(n-1, tmp_matrix)
            if tmp_value < _min:
                _min = tmp_value
        return _min
            
        
T = int(input())
for case in range(T):
    answer = f'#{case+1} '
    N, W, H = list(map(int,input().split()))
    _matrix = list()
    for i in range(H):
        row = list(map(int,input().split()))
        _matrix.append(row)
    init_sum = get_sum(_matrix)
    value = complete_search(N, _matrix)
    
    print(answer +str(value))
```

```python
#5644
import sys

sys.stdin = open("sample_4.txt","r")

global C_MATRIX
global P_MATRIX
global charge

def draw_BC(x, y, c, p, _id):
    global C_MATRIX
    global P_MATRIX
    
    # C_matrix에 BC id 적기
    n =0
    for i in range(-c, c+1):
        if 0 <= x+i < 10:
            for j in range(-n, n+1):
                if 0 <= y+j < 10:
                    C_MATRIX[x+i][y+j].add(_id)
                    P_MATRIX[x+i][y+j].add(p)
        if i < 0:
            n += 1
        else:
            n -= 1
            

def charge_A_B(loc_A, loc_B, c_p_dict, t):
    global charge
    
    bc_A = C_MATRIX[loc_A[0]][loc_A[1]]
    bc_B = C_MATRIX[loc_B[0]][loc_B[1]]

    if len(bc_A & bc_B) >= 1:
        AB = sorted([c_p_dict[bc_id] for bc_id in list(bc_A & bc_B)])
        AB_first = AB[-1]
        AB_second = 0
        if len(AB) >= 2:
            AB_second = AB[-2]
            
        A_minus_B = bc_A.difference(bc_B)
        B_minus_A = bc_B.difference(bc_A)
        
        if len(A_minus_B):
            max_A_only = max([c_p_dict[bc_id] for bc_id in list(A_minus_B)])
        if len(B_minus_A):
            max_B_only = max([c_p_dict[bc_id] for bc_id in list(B_minus_A)])
        
        if (len(A_minus_B) == 0) & (len(B_minus_A) == 0):
            charge += AB_first
            charge += AB_second
        elif (len(A_minus_B) > 0) & (len(B_minus_A) == 0):
            charge += AB_first
            charge += max(AB_second, max_A_only)
        elif (len(A_minus_B) == 0) & (len(B_minus_A) > 0):
            charge += AB_first
            charge += max(AB_second,max_B_only)
        elif (len(A_minus_B) > 0) & (len(B_minus_A) > 0):
            charge += max([(max_A_only + AB_first), (max_B_only + AB_first), (max_A_only + max_B_only), (AB_first + AB_second)])
    else:
        charge += max(P_MATRIX[loc_A[0]][loc_A[1]])
        charge += max(P_MATRIX[loc_B[0]][loc_B[1]])
    
    
# T = int(sys.stdin.readline())
T = int(input())
for i in range(T):
    answer = f"#{i+1} "
    C_MATRIX = [[0]*10 for _ in range(10)]
    P_MATRIX = [[0]*10 for _ in range(10)]
    
    for i in range(10):
        for j in range(10):
            C_MATRIX[i][j] = set()
            P_MATRIX[i][j] = set([0])
            
#     M, A = list(map(int,sys.stdin.readline().split()))
#     move_A = list(map(int,sys.stdin.readline().split()))
#     move_B = list(map(int,sys.stdin.readline().split()))
    
    M, A = list(map(int,input().split()))
    move_A = list(map(int,input().split()))
    move_B = list(map(int,input().split()))
    
    c_p_dict = dict()
    for j in range(A):
#         x, y, c, p = list(map(int,sys.stdin.readline().split()))
        x, y, c, p = list(map(int,input().split()))
        c_p_dict[j+1] = p
        draw_BC(x-1, y-1, c, p, j+1)
    
    loc_A = [0,0]
    loc_B = [9,9]
    charge = 0
    t = 0
    for ma, mb in zip(move_A, move_B):
        # 현재 위치에서 충전 가능 여부 확인 및 충전
        charge_A_B(loc_A, loc_B, c_p_dict, t)
                
        #이동
        if ma == 1:
            loc_A[1] -= 1
        elif ma == 2:
            loc_A[0] += 1
        elif ma == 3:
            loc_A[1] += 1
        elif ma == 4:
            loc_A[0] -= 1
            
        if mb == 1:
            loc_B[1] -= 1
        elif mb == 2:
            loc_B[0] += 1
        elif mb == 3:
            loc_B[1] += 1
        elif mb == 4:
            loc_B[0] -= 1
        t += 1
    
    charge_A_B(loc_A, loc_B, c_p_dict, t)
    
            
    print(answer+str(charge))
        

    
```

```python
#2112
# import sys
# sys.stdin = open("sample_5.txt","r")
from itertools import combinations, product

global MATRIX
global D, W, K

def check_pass():
    global MATRIX
    global K
    
    pass_flag = True
    for j in range(len(MATRIX[0])):
        cnt = 1
        prev_value = -1
        for i in range(len(MATRIX)):
            if MATRIX[i][j] == prev_value:
                cnt += 1
                if cnt == K:
                    break
            else:
                prev_value = MATRIX[i][j]
                cnt = 1
        if cnt < K:
            pass_flag = False
            break
    
    return pass_flag
    
# 
def change_matrix(num, rows):
    global MATRIX
    global W
    
    if len(rows) == 0:        
        return check_pass()
    else:
        tmp_row = MATRIX[rows[0]]
        MATRIX[rows[0]] = [num]*W
        pass_flag_0 = change_matrix(0, rows[1:])
        pass_flag_1 = change_matrix(1, rows[1:])
        MATRIX[rows[0]] = tmp_row
        return (pass_flag_0 | pass_flag_1)
    
    
# T = int(sys.stdin.readline())
T = int(input())
for case in range(T):
    answer = f'#{case+1} '
#     D, W, K = list(map(int, sys.stdin.readline().split()))
    D, W, K = list(map(int, input().split()))
    MATRIX = []
    for _ in range(D):
#         MATRIX.append(list(map(int, sys.stdin.readline().split())))
        MATRIX.append(list(map(int, input().split())))
    
    if check_pass():
        print(answer +str(0))
    elif K == 1:
        print(answer +str(0))
    elif K == D:
        num_0 = 0
        num_1 = 0
        for row in MATRIX:
            if sum(row) == 0:
                num_0 += 1
            elif sum(row) == W:
                num_1 += 1
        print(answer +str(D-max(num_0, num_1)))
    else:
        num_insert = K
        pass_flag = False
        for i in range(1, K):
            rows_list = list(combinations(list(range(D)), i))
            for rows in rows_list:
                pass_flag = (change_matrix(0, rows) | change_matrix(1, rows))
                if pass_flag:
                    break
            if pass_flag:
                num_insert = i
                break
        print(answer +str(num_insert))
```
